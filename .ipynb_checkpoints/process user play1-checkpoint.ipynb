```python
import pandas as pd 
```

```python
data = pd.read_json(r'C:\Users\25472\Documents\userplaysmall.dat')
```

```python
data = data[['userid', 'albumid']]
```

```python
# data
```

```python
new_df = pd.DataFrame(data['albumid'].str.split('&').tolist(), index=data['userid']).stack()
new_df = new_df.reset_index([0, 'userid'])
new_df.columns = ['userid', 'albumid']
```

```python
new_df.drop_duplicates()
```

```python
videoinfo = pd.read_json(r'C:\Users\25472\Documents\video_parsed.dat', encoding='utf-8')
videoinfo = videoinfo[['id', 'actor', 'Updatetype', 'area', 'channelid', 'cpid', 'director', 'language', 'paytype', 'score', 'tag', 'year']]
videoinfo['updatetype'] = videoinfo['Updatetype']
videoinfo = videoinfo.drop(['Updatetype'], axis=1)
```

```python


playdata_with_feature = pd.merge(new_df, videoinfo, left_on='albumid', right_on='id')
```

```python
# playdata_with_feature.drop_duplicates()
```

```python
result = playdata_with_feature[['userid']].drop_duplicates()

def create_merged(result, col_name):
    def func(line):
        line = [str(l) for l in line]
        return ','.join(line)
    actors = playdata_with_feature.groupby('userid')[col_name].apply(lambda a : func(a)).to_frame()
    result = pd.merge(result, actors, how='left', on=['userid'])
    return result
```

```python
result = create_merged(result, 'actor')
result = create_merged(result, 'director')
result = create_merged(result, 'area')
result = create_merged(result, 'channelid')
result = create_merged(result, 'cpid')
result = create_merged(result, 'tag')
result = create_merged(result, 'paytype')
result = create_merged(result, 'language')
result
```

```python
def func(l):
    try:
        a = float(l)
    except:
        a = 0
    return a

playdata_with_feature['score'] = playdata_with_feature['score'].apply(lambda l : func(l))
```

```python
def concat(pd1, pd2, new_column_name):
    result = pd.merge(pd1, pd2, how='left', on=['userid'])
    return result

score_mean = playdata_with_feature.groupby(['userid'])['score'].mean().to_frame()
score_max = playdata_with_feature.groupby(['userid'])['score'].max().to_frame()

result = concat(result, score_mean, 'score_mean')
result = concat(result, score_max, 'score_max')
result


```
