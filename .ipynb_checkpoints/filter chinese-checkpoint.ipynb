```python
import pandas as pd 

# file = open('/User/qianjay/videosmall.dat', 'r')
# file = '{' + ','.join(file) + '}'

# print(file)

data = pd.read_json("/Users/qianjay/video.dat", lines=True)
data = data[['id', 'Updatetype', 'albumname', 'area', 'channelid', 'cpid', 'actor', 'director', 'language', 'paytype',
                  'score', 'tag', 'year']]
```

```python
single_categorical = ['language', 'channelid', 'paytype', 'Updatetype', 'cpid']

for name in single_categorical:
    l = data[name].drop_duplicates().tolist()
    print('{} and len {}'.format(name, len(l)))
    m = dict(zip(l, range(len(l))))
    data[name] = data[name].apply(lambda a : m[a])

    
    
    
    
```

```python

def multi_transformer(df, col_name, seperator=[',', '|', '/', '、']):
    l = []
    data_list = df[col_name].drop_duplicates().tolist()
    for i in range(len(data_list)):
        if data_list[i] == None:
            data_list[i] = 'None'
    
    def add(s, l):
        if s == None:
            s = 'None'
        for sep in seperator:
            words = s.split(sep)
            if len(words) > 1:
                l.extend(words)
                return ','.join(words)
        l.append(s)
        return s    
    
    def trans(s, m):
        l = [str(m[word]) for word in list(s.split(',')) ]
        return ','.join(l)
        
    df[col_name + 'indexed'] = df[col_name].apply(lambda line : add(line, l))
    l = list(set(l))
    print('{} and len {}'.format(col_name, len(l)))
    m = dict(zip(l, range(len(l))))
    df[col_name] = df[col_name + 'indexed'].apply(lambda line : trans(line, m))
    df = df.drop([col_name + 'indexed'], axis=1)
    return df 

def fill_na(line):
    if line == 'null':
        return 0
    return line

multi_categorical = ['area', 'director', 'actor', 'tag']

for name in multi_categorical:
    data = multi_transformer(data, name)

data['score'] = data['score'].apply(lambda line : fill_na(line))
data['year'] = data['year'].apply(lambda line : fill_na(line))
data
```

```python
json = data.to_json(orient='records', force_ascii=False)
```
